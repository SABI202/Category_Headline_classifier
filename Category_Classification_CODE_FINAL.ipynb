```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
```

This snippet prepares your text data and labels for training a supervised classification model by:<br>
	‚Ä¢	Creating the text input feature.<br>
	‚Ä¢	Encoding the target categories as numbers.<br>
	‚Ä¢	Splitting the dataset properly into training and testing subsets.<br>

```python
X = (df['headline'] + " " + df['short_description']).fillna("")

# Encode categorical labels
le = LabelEncoder()
y = le.fit_transform(df['category_grouped'])

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

This pipeline neatly packages text vectorization and model training into a single object. It‚Äôs designed to:<br>
	‚Ä¢	Extract meaningful `TF-IDF` features from your text data.<br>
	‚Ä¢	Train a logistic regression model capable of handling multiple categories.<br>

You can now train this pipeline directly with your training data and then predict or evaluate it on new samples.

```python
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=10000, ngram_range=(1, 2))),
    ('clf', LogisticRegression(
        multi_class='multinomial',  # softmax regression
        solver='lbfgs',
        max_iter=1000,
        random_state=42
    ))
])
```

<br>This line trains the machine learning pipeline using the training data. It fits the TfidfVectorizer on the input text (X_train) and transforms it into feature vectors. Then, it trains the LogisticRegression model on the corresponding labels (y_train). The entire process is handled in one step through the pipeline.

```python
pipeline.fit(X_train, y_train)
```

<br>The trained pipeline to predict labels for the test data (X_test). The predicted labels are stored in y_pred. Then, it prints a classification report comparing the true labels (y_test) to the predictions. The report includes metrics like precision, recall, and F1-score for each category.

```python
y_pred = pipeline.predict(X_test)
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

<br>Imports the confusion_matrix function from scikit-learn to evaluate the model‚Äôs predictions. It computes the confusion matrix by comparing the true labels (y_test) with the predicted labels (y_pred). The confusion matrix shows the counts of correct and incorrect predictions for each class. Finally, it prints the confusion matrix to help assess model performance.

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

<br>
Calculates True Positives (TP), False Positives (FP), False Negatives (FN), and True Negatives (TN) from the confusion matrix. It sums these values across all classes to get total counts. Then, it prints the overall totals, including the total number of samples. This helps in understanding the detailed performance of the classification model.

```python
TP = np.diag(cm)
FP = np.sum(cm, axis=0) - TP
FN = np.sum(cm, axis=1) - TP
TN = np.sum(cm) - (TP + FP + FN)

# Total sums
total_TP = TP.sum()
total_FP = FP.sum()
total_FN = FN.sum()
total_TN = TN.sum()

print("\n=== Totals ===")
print(f"Total samples: {cm.sum()}")
print(f"True Positives (TP): {total_TP}")
print(f"False Positives (FP): {total_FP}")
print(f"False Negatives (FN): {total_FN}")
print(f"True Negatives (TN): {total_TN}")
```

<br>This code imports the pickle module to handle object serialization. It opens a file named `multiclass_logistic_regression.pkl` in write-binary mode. <br>The trained machine learning pipeline is then saved (serialized) into this file. This allows the model to be stored and loaded later without retraining.

```python
import pickle
with open('multiclass_logistic_regression.pkl', 'wb') as f:
    pickle.dump(pipeline, f)
```

```python
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset, DatasetDict
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import pandas as pd
import torch
```

```python
df['text'] = df['headline'].fillna('') + ' ' + df['short_description'].fillna('')
label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['category_grouped'])
num_classes = len(label_encoder.classes_)
```

```python
train_df, test_df = train_test_split(df[['text', 'label']], test_size=0.2, stratify=df['label'], random_state=42)
dataset = DatasetDict({
    'train': Dataset.from_pandas(train_df.reset_index(drop=True)),
    'test': Dataset.from_pandas(test_df.reset_index(drop=True))
})
```

```python
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
def tokenize_fn(example):
    return tokenizer(example['text'], padding='max_length', truncation=True, max_length=128)

dataset = dataset.map(tokenize_fn, batched=True)
dataset = dataset.remove_columns(['text'])
dataset.set_format('torch')
```

```python
model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=num_classes)
```

```python
training_args = TrainingArguments(
    output_dir='./results',
    eval_strategy='epoch',
    save_strategy='epoch',
    save_total_limit=2,
    load_best_model_at_end=True,
    metric_for_best_model='accuracy',
    greater_is_better=True,
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    weight_decay=0.1,
    logging_dir='./logs',
    logging_steps=50,
    label_smoothing_factor=0.1,
    fp16=True
)
```

```python
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted')
    acc = accuracy_score(labels, preds)
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset['train'],
    eval_dataset=dataset['test'],
    compute_metrics=compute_metrics)
```

```python

trainer.train()

#Save final model
trainer.save_model("./results/best_model")
tokenizer.save_pretrained("./results/best_model")
```

```python
import joblib

# Reuse the same label encoder you trained with
joblib.dump(label_encoder, './results/best_model/label_encoder.pkl')
```

```python
from transformers import DistilBertForSequenceClassification, DistilBertTokenizer

# Load model and tokenizer
model = DistilBertForSequenceClassification.from_pretrained("./results/best_model")
tokenizer = DistilBertTokenizer.from_pretrained("./results/best_model")

# Example: Inference on new text
text = "Breaking news: "
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)

# Run prediction
model.eval()
with torch.no_grad():
    outputs = model(**inputs)
    probs = torch.nn.functional.softmax(outputs.logits, dim=1)
    predicted_label = torch.argmax(probs, dim=1).item()

label_encoder = joblib.load('./results/best_model/label_encoder.pkl')


print("Predicted label:", label_encoder.inverse_transform([predicted_label]))
```

1.	Load the model and tokenizer:
Load the fine-tuned DistilBERT model and its corresponding tokenizer from the saved directory (./results/best_model).
2.	Prepare input text:
A sample text "Breaking news: " is tokenized, converted to input tensors with padding and truncation to a max length of 128 tokens.
3.	Run the model in evaluation mode:
Disable gradient calculations (torch.no_grad()) for efficiency during inference.
Pass the tokenized inputs to the model to get raw logits.
4.	Convert logits to probabilities:
Apply softmax to logits to obtain class probabilities.
5.	Predict the label:
Take the class with the highest probability as the predicted label index.
6.	Load the label encoder:
The saved LabelEncoder is loaded from disk to convert the predicted numeric label back to its original text category.
7.	Print the predicted label:
Outputs the readable category name corresponding to the predicted label.

```python

from transformers import DistilBertForSequenceClassification, DistilBertTokenizer

# Load model and tokenizer
model = DistilBertForSequenceClassification.from_pretrained("./results/best_model")
tokenizer = DistilBertTokenizer.from_pretrained("./results/best_model")

# Example: Inference on new text
text = "Breaking news: "
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)

# Run prediction
model.eval()
with torch.no_grad():
    outputs = model(**inputs)
    probs = torch.nn.functional.softmax(outputs.logits, dim=1)
    predicted_label = torch.argmax(probs, dim=1).item()
# Load it when needed
label_encoder = joblib.load('./results/best_model/label_encoder.pkl')


print("Predicted label:", label_encoder.inverse_transform([predicted_label]))
```
